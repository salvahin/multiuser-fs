## **Kruskal-Wallis test**

### Tests with all the combinations 

* We apply the Kruskal test to see if there are significant differences between the groups, we find that there are, but we need to know in which these differences exist.

```python
from scipy.stats import kruskal

kruskal_result = kruskal(*[group["MCC"].values for _, group in data.groupby("combination")])

print(f"Kruskal-Wallis Stadistic: {kruskal_result.statistic}")
print(f"p-value: {kruskal_result.pvalue}")
```

* When you want to make comparisons between groups (as in our case) applying independent tests increases the probability of incurring in a Type I error (that is, detecting significant differences where in reality there are none), to keep the Type I error within an established limit, there are various techniques for multiple comparisons (post-hoc)

Before applying the Tukey test, we need to check if homoscedasticity is true between our groups! One of the tests to verify this is Bartlett's test.

* $H_0$ : there are no significant differences between the variances of the groups (homoscedasticity)
* $H_{\alpha}$ : there are significant differences between the variances of the groups (heteroscedasticity)

What are we looking for? Our p-value must be greater than 0.05 so that we cannot reject the null hypothesis and have homoscedasticity!

```python
import scipy.stats as stats

groups = [group["MCC"].values for _, group in data.groupby("combination")]

stat, p_value = stats.levene(*groups)

print(f"Levene test statistic: {stat}, p-value: {p_value}")
```

* Homoscedasticity was confirm!!! And since our data is normalized, we continue with the Tukey test.

```python
import scikit_posthocs as sp

posthoc = sp.posthoc_tukey(data, val_col='MCC', group_col='combination')

# we remove the values above the main diagonal to make the plot more readable 
mask = np.triu(np.ones_like(posthoc, dtype=bool))

plt.figure(figsize=(12, 8))
sns.heatmap(posthoc, annot=True, cmap=sns.color_palette("colorblind", as_cmap=True), fmt=".3f", linewidths=0.5, mask=mask)
plt.title("Post-hoc (Tukey) Test - Differences between combinations")
plt.savefig(r"C:\Users\Coord-Personal\Documents\ITZELT\data_sciece_girlypop\proyecto_salvador\imagenes\post_hoc_tukey_test_all_data.svg")
plt.show()
```

### Test by optimizer 

```python
kruskal_results = {}

for optimizer, group in data.groupby("optimizer"):
    classifier_groups = [subgroup["MCC"].values for _, subgroup in group.groupby("classifier")]  # We group the data within the optimizer and extract the 30 MCC values ​​from each classified

    kruskal_test = kruskal(*classifier_groups)
    kruskal_results[optimizer] = (kruskal_test.statistic, kruskal_test.pvalue)


for optimizer, (statistic, pvalue) in kruskal_results.items():
    print(f"Optimizer: {optimizer} - Stadistic: {statistic:.2f},  p-value: {pvalue:.4f}")
```

```python
df_kw_optimizer = pd.DataFrame(kruskal_results, index=["statistic", "p-value"]).T
df_kw_optimizer = df_kw_optimizer.reset_index().rename(columns={'index': 'optimizer'})
df_kw_optimizer.to_latex(index=False,
                  formatters={"name": str.upper},
                  float_format="{:.4f}".format)
```

## **Feature Extraction**

```python
data
```

* we have to change the data type from str to list 

```python
data["features"] = data["features"].astype(str).str.strip("[]").str.replace("'", "").str.split(", ")
```

+ Here what we do is that we join the 30 lists of each combination into a single list

```python
df_with_features = data.groupby("combination").agg({
    "features": lambda x: sum(x,[]) # joins the 30 lists of a combination into one, the sum function joins the lists
})
```

```python
df_with_features
```

* We get the unique characteristics to use as a column in our new table. What we do is use the sum function to concatenate the lists in each row into a single list, while set converts that list into a set that eliminates duplicate values.

```python
unique_features = set((df_with_features["features"].sum()))
unique_features
```

```python
len(unique_features)
```

* We create a null df with the combinations and the feature columns

```python
feature_counts = pd.DataFrame(columns=["combination"] + list(unique_features))
```

```python
from collections import Counter

for i, row in df_with_features.iterrows():
    conteo = Counter(row["features"])  # Contar ocurrencias de cada feature
    datos_fila = {feature: conteo.get(feature, 0) for feature in unique_features}  # Obtener los conteos
    datos_fila.update({"combination": row.name})  # Usar 'row.name' para obtener el índice de la combinación
    feature_counts = pd.concat([feature_counts, pd.DataFrame([datos_fila])], ignore_index=True)
```

* We have to make sure that the indexes are the combinations to be able to make our heatmap, otherwise it can´t be done

```python
feature_counts.set_index("combination", inplace = True)
```

```python
feature_counts
```

* convert the values to float

```python
feature_counts = feature_counts.astype(float)
```

* order the number of features from highest to lowest so that you can see which ones were chosen the most.

```python
feature_counts = feature_counts[feature_counts.sum().sort_values(ascending=False).index]
```

* now we can see our matrix to compare the number of features selected by combination of optimizer and classifier

```python
plt.figure(figsize = (16,9))
sns.heatmap(feature_counts, vmin = 0, vmax = 30, cmap = sns.color_palette("colorblind", as_cmap = True) , annot = True)
plt.savefig(r"C:\Users\Coord-Personal\Documents\ITZELT\data_sciece_girlypop\proyecto_salvador\imagenes\heatmap_features_all_data.svg")
plt.show()
```

```python
plt.figure(figsize=(18,9))
ax = sns.barplot(x=feature_counts.sum(), y=feature_counts.columns, orient="h", palette="colorblind")

for container in ax.containers:
    ax.bar_label(container, fontsize=10, fmt="%.0f")

plt.xlabel("Frecuency")
plt.ylabel("Features")
plt.title("Features frecuency by combination")
plt.savefig(r"C:\Users\Coord-Personal\Documents\ITZELT\data_sciece_girlypop\proyecto_salvador\imagenes\features_frecuency_all_data.svg")
plt.show()
```
